```python
dataarray = xr.open_rasterio(path+ "//" + Imagenes[10])
df = dataarray[0].to_pandas()
df = df.stack()
df = pd.DataFrame(df)
df = df.reset_index()
Lon = np.tile(df["x"],85)
Lat = np.tile(df["y"],85)
```

```python
def crearDF(ruta, listImg):
    dfl = []
    for i in range(len(listImg)):
        dataarray = xr.open_rasterio(ruta+ "//" + listImg[i])
        df = dataarray[0].to_pandas()
        df = df.stack()
        df = pd.DataFrame(df)
        #df = df.append(df, ignore_index=False)
        #df = pd.concat([df])
        dfl.append(df[0].to_list())
    return dfl
```

```python
VHI = crearDF(path,Imagenes)
```

```python
VHI
```

```python
VHI = list(map(float, chain.from_iterable(VHI)))
```

```python
VHIdf = {"VHI":VHI}
VHIdf = pd.DataFrame(VHIdf)
VHIdf
```

```python
len(VHIdf)
```

```python
VHIdf["x"] = Lon
```

```python
VHIdf["y"] = Lat
```

```python
VHIdfG = gpd.GeoDataFrame(VHIdf, geometry=gpd.points_from_xy(VHIdf.x, VHIdf.y),crs="+proj=sinu +lon_0=0 +x_0=0 +y_0=0 +datum=WGS84 +units=m")
```

```python
VHIdfG
```

```python
VHIdfG = VHIdfG.to_crs("epsg:4326")
```

```python
DataVHI = pd.DataFrame(VHIdfG["geometry"].x)
```

```python
DataVHI["y"] = pd.DataFrame(VHIdfG["geometry"].y)
```
