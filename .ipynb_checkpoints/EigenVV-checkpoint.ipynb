# **Eigen vectores y eigen valores**

<p>Dada una matriz cuadrada $A$, un eigen vector ($v$) y valor ($\lambda$) quedas definidos como:</p><br>

$A v = \lambda v$

<p>Para estimar los eigen vectores y valores primero debemos definir una matriz identidad ($I$) para obtener las mismas dimensiones de la matriz $A$: </p><br>

$A v = \lambda I v$<br>
$A v - \lambda I v = 0$<br>
$\mid A - \lambda I\mid v = 0$

<p>Un ejemplo de lo anteriormente descrito es: </p><br>
$ A = \begin{bmatrix}
-6 & 3 \\
4 & 5
\end{bmatrix}$ <br>
$\begin{bmatrix}
-6 & 3 \\
4 & 5
\end{bmatrix}-\lambda \begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix} = 0$

$\begin{bmatrix}
-6 & 3 \\
4 & 5
\end{bmatrix}- \begin{bmatrix}
\lambda & 0 \\
0 & \lambda
\end{bmatrix} = 0$

$\begin{bmatrix}
-6 - \lambda & 3 \\
4 & 5 - \lambda
\end{bmatrix}= 0$

<p>Por determinantes obtenemos: </p><br>

$(-6 -\lambda)(5 - \lambda)-(3)(4)$

$\lambda^{2} - \lambda -42 = 0$

<p>Por ecuación cuadrática obtenemos que: </p><br>
$\lambda = -7$<br>
$\lambda = 6$

<p>Cuando $\lambda$ es 6</p>

$\begin{bmatrix}
-6 & 3 \\
4 & 5
\end{bmatrix}\begin{bmatrix}
x \\
y
\end{bmatrix} = 6 \begin{bmatrix}
x \\
y
\end{bmatrix}$

$-6x + 3y = 6x$

$4x + 5y = 6y$

$-12x + 3y = 0$

$4x - y = 0$

$x = 1$ <br>
$y = 4$

$\begin{bmatrix}
-6 & 3 \\
4 & 5
\end{bmatrix}\begin{bmatrix}
1 \\
4
\end{bmatrix} =  \begin{bmatrix}
6 \\
24
\end{bmatrix}$

$ 6 \begin{bmatrix}
1 \\
4
\end{bmatrix} = \begin{bmatrix}
6 \\
24
\end{bmatrix}$

<p>Cuando $\lambda$ es -7</p>

$\begin{bmatrix}
-6 & 3 \\
4 & 5
\end{bmatrix}\begin{bmatrix}
x \\
y
\end{bmatrix} = -7 \begin{bmatrix}
x \\
y
\end{bmatrix}$

$-6x + 3y = -7x$

$4x + 5y = -7y$

$x + 3y = 0$

$4x + 12y =0$

$x = -3$ <br>
$y = 1$

$\begin{bmatrix}
-6 & 3 \\
4 & 5
\end{bmatrix}\begin{bmatrix}
-3 \\
1
\end{bmatrix} =  \begin{bmatrix}
21 \\
-7
\end{bmatrix}$

$ -7 \begin{bmatrix}
-3 \\
1
\end{bmatrix} = \begin{bmatrix}
21 \\
-7
\end{bmatrix}$

```python
plt.scatter([-6,4],[3,5])
plt.scatter([1],[4], color="red")
plt.scatter([-3],[1], color="red")
plt.annotate("(-6,3)",(-5.9,3.1), color="blue")
plt.annotate("(4,5)",(3.9,5.1), color="blue")
plt.annotate("(1,4)",(0.9,4.1), color="red")
plt.annotate("(-3,1)",(-2.9,1.1), color="red")
plt.ylabel("y")
plt.xlabel("x")
plt.title("Eigen Vector $v$ y matriz A")
plt.show()
```

```python
plt.scatter([6],[24], color="purple")
plt.scatter([21],[-7], color="purple")
plt.annotate("(6,24)",(5.9,24.1), color="purple")
plt.annotate("(21,-7)",(19.9,-6.1), color="purple")
plt.ylabel("y")
plt.xlabel("x")
plt.title("Eigen Vectores por matriz A ($A v$)")
plt.show()
```

```python
plt.scatter([-6,4],[3,5])
plt.scatter([6],[24], color="purple")
plt.scatter([21],[-7], color="purple")
plt.annotate("(6,24)",(5.9,24.1), color="purple")
plt.annotate("(21,-7)",(19.9,-6.1), color="purple")
plt.annotate("(-6,3)",(-5.9,3.1), color="blue")
plt.annotate("(4,5)",(3.9,5.1), color="blue")
plt.ylabel("y")
plt.xlabel("x")
plt.title("$Av$ y matriz A")
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter([-6,4],[3,5])
plt.arrow(0,0,4,5, color="blue")
plt.arrow(0,0,-6,3, color="blue")
plt.arrow(-6,3,10,2, color="blue")
plt.scatter([6],[24], color="purple")
plt.arrow(0,0,6,24, color="purple")
plt.scatter([1],[4], color="red")
plt.arrow(0,0,1,4, color="red")
plt.scatter([21],[-7], color="purple")
plt.arrow(0,0,21,-7, color="purple")
plt.scatter([-3],[1], color="red")
plt.arrow(0,0,-3,1, color="red")
plt.arrow(-3,1, 4,3, color="red")
plt.arrow(21,-7, -15,31, color="purple")
plt.ylabel("y")
plt.xlabel("x")
plt.annotate("v (1,4)",(-2,5), color="red")
plt.annotate("v (-3,1)",(-3,-2), color="red")
plt.annotate("Av (6,24)",(1.5,23), color="purple")
plt.annotate("Av (21,-7)",(15,-4), color="purple")
plt.annotate("Matriz A",(2.5,2), color="blue")
plt.annotate("Eigen vectores",(0,-2), color="red")
plt.annotate("Producto $Av$",(15.5,5), color="purple")
plt.show()
```

<p>De acuerdo con la gráfica mostrada anteriormente, la matriz podría girar en dirección al trayecto (6,24) denotado por el vector propio correspondien al primer eigen valor ($\lambda = 6$), o podría girar en dirección al trayecto (21,7) denotado por el vector propio del segundo eigen valor  ($\lambda = -7$). Observen la siguiente imagen, la cual gira en sentido del producto de su vector propio con la matriz que lo compone:</p><br>

```python
a = np.array([[2, 2, 4], 
              [1, 3, 5],
              [2, 3, 4]])
```

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
x=[2,1,2]
y=[4,3,3]
z=[4,5,4]
ax.plot3D(x, y, z, 'blue')
ax.scatter3D(x, y, z, c=z, cmap='hsv')
plt.show()
```

```python
w,v=eig(a)
print('E-value:', w)
print('E-vector', v)
```

<p>Los eigen vectores son vectores que no cambian su orientación, sino que solo se escalan por un factor de su eigen valor  correspondiente.</p>

<p> No importa qué tipo de matriz de transición "A" tenga, si se logra encontrar sus valores propios y vectores propios, la transición que usa la matriz "A" en los vectores propios no cambia su dirección, sino que solo se escala por un factor de los valores propios correspondientes.
